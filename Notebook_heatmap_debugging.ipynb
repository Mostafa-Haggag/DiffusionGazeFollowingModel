```python
import numpy as np 
import scipy.stats as stats

hist1, _ = np.histogram(df_3.to_numpy(), bins=100)
hist2, _ = np.histogram(df_3v.to_numpy(), bins=100)

# Use a histogram-based similarity measure like Earth Mover's Distance (EMD)
emd = stats.wasserstein_distance(hist1, hist2)
print("Earth Mover's Distance (EMD):", emd)# a lower value is better.
```

```python
from sklearn.metrics import pairwise_distances

cosine_similarity = 1 - pairwise_distances(df_3.to_numpy().reshape(1, -1), df_3v.to_numpy().reshape(1, -1), metric="cosine")[0, 0]
print("Cosine Similarity:", cosine_similarity)
# . Therefore, a higher cosine similarity value signifies that the two vectors are more aligned and have a greater degree of similarity,
# whereas a lower value suggests greater dissimilarity or a larger angle between the vectors.
```

```python
mse = np.mean((df_3.to_numpy() - df_3v.to_numpy()) ** 2)
rmse = np.sqrt(mse)
print("MSE:", mse)
print("RMSE:", rmse)
```

```python
correlation = np.corrcoef(df_3.to_numpy().flatten(), df_3v.to_numpy().flatten())[0, 1]
print("Correlation Coefficient:", correlation)
```
